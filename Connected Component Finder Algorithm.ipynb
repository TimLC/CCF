```
import datetime
```

```
linesFile = sc.textFile("./web_Google.txt")
edges = linesFile.filter(lambda x: x[0] != "#").map(lambda x : (int(x.split()[0]), int(x.split()[1])))
```

```
def comparingEdges(pair):
  valueList = []
  edgeList = []
  key, values = pair
  minValue = key
  
  for value in values:
    if value < minValue:
       minValue = value
    valueList.append(value)
  if minValue < key:
    edgeList.append((key, minValue))
    for value in valueList:
      if minValue != value:
        newPair.add(1)
        edgeList.append((value, minValue))
  return edgeList
```

```
numberIteration = 0
newPair = sc.accumulator(0)
newPairCreate = True

start = datetime.datetime.now()

while newPairCreate == True:
  numberIteration += 1
  newPair.value = 0
  
  # CCF-Iterate
  jopProcessing = edges.flatMap(lambda x: (x, (x[1], x[0]))).groupByKey().flatMap(lambda x: comparingEdges(x)).sortByKey()
  
  # CCF-Dedup
  edges = jopProcessing.distinct()
  
  # Conditionnel test
  if newPair.value == 0:
    newPairCreate = False
    
  print("Iteration: " + str(numberIteration) + " | number of pair : " + str(newPair.value))
  
duration = datetime.datetime.now() - start
print("Time execution : " + str(duration))
```
